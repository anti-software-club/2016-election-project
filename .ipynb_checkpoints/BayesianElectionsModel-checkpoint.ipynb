```python
# Library to for CSV file streams
import csv

# Various data structures for storing demographic +
# past voting history data.
age_dist = []
region_dist = []
income_dist = []
race_dist = []
past_training_data = []
stat_age_dist = []
stat_race_dist = []
stat_reg_dist = []
stat_inc_dist = []
dem_state_ratios = {}

# Structure for final results and predictions for all states.
final_res = {}
total_dem_ecvotes = 0
total_rep_ecvotes = 0

# Various CSV reading stream methods, storing each
# category into its respective dictionary distribution
with open('trainingstatesdata.csv', 'rU') as file:
    training_data = csv.DictReader(file)
    for row in training_data:
        past_training_data.append(row)

with open('race.csv', 'rU') as file:
    race_data = csv.DictReader(file)
    for row in race_data:
        race_dist.append(row)

with open('income.csv', 'rU') as file:
    inc_data = csv.DictReader(file)
    for row in inc_data:
        income_dist.append(row)

with open('region.csv', 'rU') as file:
    reg_data = csv.DictReader(file)
    for row in reg_data:
        region_dist.append(row)

with open('age.csv', 'rU') as file:
    age_data = csv.DictReader(file)
    for row in age_data:
        age_dist.append(row)

with open('static-age.csv', 'rU') as file:
    stat_age_data = csv.DictReader(file)
    for row in stat_age_data:
        stat_age_dist.append(row)

with open('static-race.csv', 'rU') as file:
    stat_race_data = csv.DictReader(file)
    for row in stat_race_data:
        stat_race_dist.append(row)

with open('static-region.csv', 'rU') as file:
    stat_reg_data = csv.DictReader(file)
    for row in stat_reg_data:
        stat_reg_dist.append(row)
        
with open('static-income.csv', 'rU') as file:
    stat_inc_data = csv.DictReader(file)
    for row in stat_inc_data:
        stat_inc_dist.append(row)

# Compiles percentage of democrat wins from 1992-2008
# in all six elections, with greater weight on 2008 and
# 2012 elections.
for res in past_training_data:
    dem_state_ratios[res['State']] = 0
    count = 0.0
    for field in training_data.fieldnames:
        if (field != "State" and field != ""):
            if (field[0:4] == "2008" or field[0:4] == "2012"):
                count = count + int(res[field]) * .25
            else:
                count = count + int(res[field]) * .125
    if (count == 1.0):
        dem_state_ratios[res['State']] = .97
    elif (count == 0.0):
        dem_state_ratios[res['State']] = .03
    else:
        dem_state_ratios[res['State']] = count

# Creates general mapping for electoral college votes + race distribution
ev_map = {}
for elem in race_dist:
    if (elem['State'] != 'United States'):
        ev_map[elem['State']] = int(elem['Electoral Votes'])

# Creates general mapping between states and states racial distribution.
racial_map = {}
for elem in race_dist:
    if (elem['State'] != 'United States'):
        ev_map[elem['State']] = elem

def race_coeff(state, party):
#    race_coeff = 0.0
#    ra_coeff = race_dist[]
    pass

def age_coeff(state, party):
#    age_coeff = 0.0
    pass

# Creates general mapping between income brackets and associated information.
income_map = {}
for elem in stat_inc_dist:
    if (elem['Income Bracket'] == "< $15,000"):
        income_map['bracket1'] = elem;
    elif (elem['Income Bracket'] == "$15,000-29,999"):
        income_map['bracket2'] = elem;
    elif (elem['Income Bracket'] == "$30-49,999"):
        income_map['bracket3'] = elem;
    elif (elem['Income Bracket'] == "$50-74,999"):
        income_map['bracket4'] = elem;
    elif (elem['Income Bracket'] == "$75,000-100,000"):
        income_map['bracket5'] = elem;
    elif (elem['Income Bracket'] == "$100,000+"):
        income_map['bracket6'] = elem;
    

def income_coeff(state, party):
    coeff = 1.0
    for elem in income_dist:
        if(elem['State'] == state):
            # Median income group affects which income brackets are weighted higher
            if (elem['Median Income Group'] == '1'):
                coeff = coeff * float(income_map['bracket1']['Average %']) * .24
                coeff = coeff * float(income_map['bracket2']['Average %']) * .33
                coeff = coeff * float(income_map['bracket3']['Average %']) * .46
                coeff = coeff * float(income_map['bracket4']['Average %']) * .27
                coeff = coeff * float(income_map['bracket5']['Average %']) * .18
                coeff = coeff * float(income_map['bracket6']['Average %']) * .12
            elif (elem['Median Income Group'] == '2'):
                coeff = coeff * float(income_map['bracket1']['Average %']) * .19
                coeff = coeff * float(income_map['bracket2']['Average %']) * .27
                coeff = coeff * float(income_map['bracket3']['Average %']) * .49
                coeff = coeff * float(income_map['bracket4']['Average %']) * .35
                coeff = coeff * float(income_map['bracket5']['Average %']) * .19
                coeff = coeff * float(income_map['bracket6']['Average %']) * .13
            elif (elem['Median Income Group'] == '3'):
                coeff = coeff * float(income_map['bracket1']['Average %']) * .16
                coeff = coeff * float(income_map['bracket2']['Average %']) * .25
                coeff = coeff * float(income_map['bracket3']['Average %']) * .41
                coeff = coeff * float(income_map['bracket4']['Average %']) * .37
                coeff = coeff * float(income_map['bracket5']['Average %']) * .22
                coeff = coeff * float(income_map['bracket6']['Average %']) * .14         
            elif (elem['Median Income Group'] == '4'):
                coeff = coeff * float(income_map['bracket1']['Average %']) * .13
                coeff = coeff * float(income_map['bracket2']['Average %']) * .21
                coeff = coeff * float(income_map['bracket3']['Average %']) * .33
                coeff = coeff * float(income_map['bracket4']['Average %']) * .43
                coeff = coeff * float(income_map['bracket5']['Average %']) * .24
                coeff = coeff * float(income_map['bracket6']['Average %']) * .15    
            if (party == 'D'):
                coeff = coeff * float(income_map['bracket1']['Average P(D | income i)'])
                coeff = coeff * float(income_map['bracket2']['Average P(D | income i)'])
                coeff = coeff * float(income_map['bracket3']['Average P(D | income i)'])
                coeff = coeff * float(income_map['bracket4']['Average P(D | income i)'])
                coeff = coeff * float(income_map['bracket5']['Average P(D | income i)'])
                coeff = coeff * float(income_map['bracket6']['Average P(D | income i)'])
            else:
                coeff = coeff * float(income_map['bracket1']['Average P(R | income i)'])
                coeff = coeff * float(income_map['bracket2']['Average P(R | income i)'])
                coeff = coeff * float(income_map['bracket3']['Average P(R | income i)'])
                coeff = coeff * float(income_map['bracket4']['Average P(R | income i)'])
                coeff = coeff * float(income_map['bracket5']['Average P(R | income i)'])
                coeff = coeff * float(income_map['bracket6']['Average P(R | income i)'])
    return coeff

def region_coeff(state, party):
    reg_coeff = 1.0
    for elem in region_dist:
        if (elem['State'] == state):
            for reg in stat_reg_dist:
                if (elem['Region'] == reg['Region']):
                    if (party == 'D'):
                        reg_coeff = reg_coeff * float(reg['Average P(D | region r)'])
                        reg_coeff = reg_coeff * float(reg['Average %'])
                    else:
                        reg_coeff = reg_coeff * float(reg['Average P(R | region r)'])
                        reg_coeff = reg_coeff * float(reg['Average %'])
                    break
            break
    return reg_coeff

def get_likelihood(state, party):
    if (party == 'D'):
        likelihood = dem_state_ratios[state]
#         likelihood = likelihood * race_coeff(state, 'D')
#         likelihood = likelihood * age_coeff(state, 'D')
        likelihood = likelihood * region_coeff(state, 'D')
        likelihood = likelihood * income_coeff(state, 'D')
        return float(likelihood)
    elif (party == 'R'):
        likelihood = (1 - dem_state_ratios[state])
#         likelihood = likelihood * race_coeff(state, 'R')
#         likelihood = likelihood * age_coeff(state, 'R')
        likelihood = likelihood * region_coeff(state, 'R')
        likelihood = likelihood * income_coeff(state, 'R')
        return float(likelihood)

for state in dem_state_ratios.keys():
    d_likelihood = get_likelihood(state, 'D')
    r_likelihood = get_likelihood(state, 'R')
    if (d_likelihood >= r_likelihood):
        final_res[state] = 'Democrat'
#        total_dem_ecvotes = total_dem_ecvotes + ev_map[state]
        print(ev_map[state])
    else:
        final_res[state] = 'Republican'
#        total_rep_ecvotes = total_rep_ecvotes + ev_map[state]
        print(ev_map[state])

print(final_res)
print(racial_map)
# print(total_dem_ecvotes)
# print(total_rep_ecvotes)
```
